1.	Apply PCA to reduce the high-dimensional performance metrics to 2 components.
2.	Plot the PCA results using a scatter plot, coloring points based on job roles

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
#TO READ CSV FILE
data = pd.read_csv("/content/employee_performance.csv")
```

```python
#TO STANDARDIZE DATASET
scaler = StandardScaler()
data_scaled=scaler.fit_transform(data)
data
```

```python
#PERFORMING PCA
pca=PCA(n_components=2)
principle_compnents=pca.fit_transform(data_scaled)
data_scaled
```

```python
#TO CONVERT IN DATAFRAME
df=pd.DataFrame(principle_compnents,columns=['PC1','PC2'])
df
```

```python
#TO PLOT SCATTER PLOT
plt.figure(figsize=(10,10))
plt.scatter(df['PC1'],df['PC2'])
plt.xlabel('principal compnent 1')
plt.ylabel('principal compnent 2')
plt.title('Principle Component Analysis')
plt.grid()
plt.show()
```
